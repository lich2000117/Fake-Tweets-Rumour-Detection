```python
import json
import pandas as pd
import glob
import os
```

```python
TWEET_DATA_PATH = "./id_data/tweet-objects/"
ID_PATH = "./id_data/test.data.txt"
SOURCE_STORY_ONLY=False
```

```python
total_id_list = []
with open(ID_PATH, "r") as f:
    for line in f:
        line = line.strip() # remove next line
        if SOURCE_STORY_ONLY:
            line = line.split(',')[0] # split into list\
        else:
            line = line.split(',')
        total_id_list.append(line)
if not SOURCE_STORY_ONLY:
    total_id_list = [item for sublist in total_id_list for item in sublist] # Flat into a single list
len(total_id_list)
```

```python
total_id_list = []
with open(ID_PATH, "r") as f:
    for line in f:
        line = line.strip().split(',') # remove next line
        source_id = line[0]
        if len(line) > 1:
            # if we have replies id
            replies_id = line[1:]
        else:
            replies_id = []
        row = [source_id, replies_id]
        total_id_list.append(row)
len(total_id_list)
```

```python
file_list = os.listdir( TWEET_DATA_PATH )
len(file_list)
```

```python
total_js = []
for js in file_list:
    with open (TWEET_DATA_PATH + js) as f:
        data = json.load(f)
        total_js.append(data)
```

```python
with open ("./full_data/test_full_v2.json", "w") as f:
    json.dump(total_js, f, indent=2)
```
